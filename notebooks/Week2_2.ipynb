```python
temperatures = [298, 320, 310]
yields = [85, 60, 90]

for i in range(3):  # rangeëŠ” ië¼ëŠ” ìˆ«ìë¥¼ ì…ë ¥í•˜ë©´ 0ë¶€í„° ì‹œì‘í•˜ì—¬ i-1ê¹Œì§€ ìˆëŠ” listë¥¼ ë§Œë“­ë‹ˆë‹¤.
  print(i) # í˜„ì¬ i ê°€ ë¬´ì—‡ì¸ì§€ í™•ì¸í•˜ê¸°!
  if yields[i] > 80:
      print("T =", temperatures[i], ": good result")
  else:
      print("T =", temperatures[i], ": bad result")
```
