### list와 dictionary의 차이

지금까지 배운 list와  
오늘 새로 등장한 dictionary는  
모두 여러 값을 한 번에 저장할 수 있습니다.

하지만 **중요한 차이**가 있습니다.

- list는 **순서(order)** 가 중요합니다
- dictionary는 **의미(meaning)** 가 중요합니다

아래 두 코드는  
같은 실험 조건을 서로 다른 방식으로 저장한 것입니다.

- 만약 순서가 바뀌면, 코드는 에러를 낼까요?

반면에,
- exp_dict에서는 각 값의 의미가 바로 드러납니다.

- dictionary는 **의미가 중요한 데이터**를 다룰 때 적합합니다.

화학 실험 조건, 샘플 정보, 분자 특성처럼  
“이 값이 무엇을 의미하는지”가 중요한 경우에는

### 딕셔너리 리스트 (Chemical Inventory)

실제 데이터는 여러 개의 딕셔너리가 리스트로 묶인 형태가 많습니다.

```python
inventory = [
    {"name": "Water", "formula": "H2O", "stock": 500},
    {"name": "Ethanol", "formula": "C2H5OH", "stock": 200},
    {"name": "Acetone", "formula": "CH3COCH3", "stock": 150}
]

for chemical in inventory:
    print(chemical["name"], "재고:", chemical["stock"], "ml")
```

각 dictionary는  
- 하나의 실험
- 하나의 샘플
- 하나의 데이터 포인트

를 의미합니다.

이런 dictionary가 수십 개, 수백 개 모이면  
그것이 바로 **머신러닝에서 말하는 데이터셋**입니다.


## if / else란?

실험 데이터를 다루다 보면  
항상 같은 계산만 하는 것이 아니라,

- 결과가 좋은지 나쁜지 판단하거나
- 조건에 따라 다른 행동을 하거나

해야 하는 경우가 많습니다.

이럴 때 사용하는 것이 **if / else**입니다.

👉 if / else는  
“조건에 따라 다른 코드를 실행하라”는 의미입니다.


```python
# 수율에 따라 실험 결과를 판단해봅니다
if experiment["yield"] > 80:
    print("Good experiment")
else:
    print("Needs improvement")
```

위 코드의 의미는 다음과 같습니다.

- 만약 (if) 수율이 80보다 크면  
  → "Good experiment" 출력
- 그렇지 않으면 (else)  
  → "Needs improvement" 출력

중요한 점:
- if 뒤에는 **조건**이 옵니다
- 조건의 결과는 True 또는 False 입니다
- if 줄 끝에는 반드시 `:` 가 있어야 합니다
- else 아래의 코드는 들여쓰기 되어야 합니다


```python
# 수율 값을 바꿔서 다시 실행해보세요
experiment["yield"] = 65

if experiment["yield"] > 80:
    print("Good experiment")
else:
    print("Needs improvement")
```

생각해봅시다.

- 같은 코드인데,
- 데이터 값만 바꿨을 뿐인데,
- 결과가 달라졌습니다.

이것이 바로
**데이터를 기반으로 판단하는 코드**입니다.

머신러닝 모델도  결국은 이런 판단을 자동으로 수행합니다.

### 다중 조건문 (if-elif-else)

pH 값에 따라 산성, 중성, 염기성을 나누어 봅시다.

```python
ph_values = [2.0, 7.0, 12.5]

for ph in ph_values:
    if ph < 7:
        print("pH", ph, ": Acidic (산성)")
    elif ph == 7:
        print("pH", ph, ": Neutral (중성)")
    else:
        print("pH", ph, ": Basic (염기성)")
```

### if / else의 결과는 사람이 읽을 수 있어야 합니다

코드의 목적은
컴퓨터를 만족시키는 것이 아니라,
사람에게 의미 있는 정보를 주는 것입니다.

아래 코드처럼 list 내에 한 개의 값에 대해서 판단을 하거나


```python
temperatures = [298, 310, 330]

if temperatures[0] > 300:
    print("First experiment is high T")
else:
    print("First experiment is low T")
```

for 문을 사용하여 여러 실험을 하나씩 검사하여,

조건에 맞는 경우/아닌 경우를 나눌 수도 있습니다.

```python
temperatures = [298, 310, 330]

for T in temperatures:
    if T > 300:
        print(T, "K : High temperature")
    else:
        print(T, "K : Low temperature")
```

두 개 이상의 list를 연동하여 확인할 수도 있습니다

```python
temperatures = [298, 320, 310]
yields = [85, 60, 90]

for i in range(3):  # range는 i라는 숫자를 입력하면 0부터 시작하여 i-1까지 있는 list를 만듭니다.
  print(i) # 현재 i 가 무엇인지 확인하기!
  if yields[i] > 80:
      print("T =", temperatures[i], ": good result")
  else:
      print("T =", temperatures[i], ": bad result")
```

list 내에 있는 문자열도 판단할 수 있습니다.

```python
solvents = ["water", "ethanol", "water"]

for s in solvents:
    if s == "water":
        print("Aqueous system")
    else:
        print("Non-aqueous system")
```

### 에너지 계산 함수 실습

빛의 파장($\lambda$)을 입력하면 에너지($E = hc / \lambda$)를 계산해주는 함수를 만들어 봅시다.
(단순화를 위해 $hc = 1240$ eV·nm 상수를 사용합니다)

```python
def calculate_energy(wavelength_nm):
    hc = 1240  # eV * nm
    energy = hc / wavelength_nm
    return energy

w1 = 400  # nm (Blue)
w2 = 700  # nm (Red)

print("400nm 에너지:", calculate_energy(w1), "eV")
print("700nm 에너지:", calculate_energy(w2), "eV")
```

```python
def check_yield(exp):
    if exp["yield"] > 80:
        print("Good experiment")
    else:
        print("Needs improvement")
```

### 함수는 계산 규칙을 저장하는 방법입니다

지금까지는
- 숫자를 직접 계산하거나
- if / else로 판단을 했습니다.

함수를 사용하면  
**자주 쓰는 계산 규칙**을  
하나의 이름으로 저장할 수 있습니다.

```python
def celsius_to_fahrenheit(C):
    F = C * 9 / 5 + 32
    return F

def celsius_to_kelvin(C):
    K = C + 273.15
    return K
```

```python
print("F: ", celsius_to_fahrenheit(25))
print("K: ", celsius_to_kelvin(25))
```

함수 하나에서  
여러 값을 계산해서  
한 번에 출력할 수도 있습니다.


```python
def celsius_to_F_and_K(C):
    F = C * 9 / 5 + 32
    K = C + 273.15
    return F, K
```

### 지수 함수 예시: Arrhenius 식

반응 속도 상수 k는  
온도에 대해 지수적으로 변합니다.

k = A · exp(-Ea / (R · T))


```python
import math  # 지수나 로그를 사용하기 위해 library를 불러온다

def arrhenius_rate(A, Ea, T):
    R = 8.314  # J/mol·K
    k = A * math.exp(-Ea / (R * T))
    print("Rate constant k:", k)
```

```python
arrhenius_rate(1e12, 50000, 300)
arrhenius_rate(1e12, 50000, 350)
```
