### 몰 농도 (Molarity) 계산하기

몰 농도($M$)는 용액 1L에 녹아 있는 용질의 몰수(mol)를 의미합니다.
공식: $M = n / V$
- $n$: 용질의 몰수 (mol)
- $V$: 용액의 부피 (L)

````python
moles = 0.25      # mol
volume_L = 0.5    # L

molarity = moles / volume_L
print("몰 농도 (M):", molarity)
````

## 3. for문 – 반복 작업하기

for문은 리스트 안의 값을 하나씩 꺼내서 반복 작업을 할 때 사용합니다.

기본 형태:
```python
for 항목 in 리스트:
    실행할 코드
```

주의:

반드시 : 를 써야 합니다

다음 줄은 반드시 들여쓰기 해야 합니다.


## 왜 for문이 필요한가?

리스트 길이가 늘어나면?
- 직접 하나씩 쓰는 것은 불가능
- "반복"이 필요함

for문 = "리스트 안의 모든 값에 대해 같은 작업 수행"


````python
# for문 없이 (비효율적)
values = [1, 2, 3, 4]

print(values[0])
print(values[1])
print(values[2])
print(values[3])
````

````python
# for문 사용 (권장)
for v in values:
    print(v)
````

````python
# for문 예제
for element in elements:
    print("원소:", element)
````

````python
temperatures = [25, 30, 40]
for T in temperatures:
    print(T, "°C")
````

## 5. 집합(set) – 중복 제거와 고유값

집합은 중복을 허용하지 않는 데이터 구조입니다.
화학에서는 복잡한 혼합물이나 분자에서 **어떤 종류의 원소가 포함되어 있는지** 알고 싶을 때 사용합니다.

````python
mixture = ["H", "O", "H", "C", "O", "H"]
unique_elements = set(mixture)

print("혼합물 원소:", mixture)
print("고유 원소 집합:", unique_elements)
````

## 6. while문 – 조건이 만족되는 동안 반복

for문은 정해진 횟수만큼 반복하지만, while문은 **조건이 참(True)인 동안** 계속 반복합니다.

````python
# 간단한 적정 시뮬레이션
solution_volume = 50  # ml
drop_size = 10        # ml

while solution_volume > 0:
    print("현재 용량:", solution_volume, "ml")
    print("10ml를 덜어냅니다.")
    solution_volume = solution_volume - drop_size

print("용액을 모두 사용했습니다.")
````
